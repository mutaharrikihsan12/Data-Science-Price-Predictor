```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('clean_car_fix.csv')
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df.info()
```

```python
df[df['Make']=='Audi']
```

# Investigate Number of Cars Distribution in Every Year

```python
#membuat dictionary dengan range tahun sesuai dari index minimal tahun
year_dic = {1990+i : sum(df['Year']==1990+i) for i in range(28)}
x_year = [1990 + i for i in range(28)]
y_year = [year_dic[1990 + i] for i in range(28)]

# Visualization
sns.set(style="whitegrid", context="talk")
f, ax1 = plt.subplots(1,1,figsize=(15,5))
ax1.set_title('Car year distribution')
sns.barplot(x=x_year, y=y_year, palette="OrRd", ax=ax1)
ax1.set_xlabel('Year')
ax1.set_ylabel('Cars')
locs, labels = plt.xticks() 
plt.xticks([0,5,10,15,20,25],['1990', '1995', '2000', '2005', '2010', '2015'])
plt.grid(axis = 'x')
plt.show()
```

# Investigate Percentage of Brand Distribution

```python
# Percentage of car per brand
counts = df['Make'].value_counts()*100/sum(df['Make'].value_counts())

# 10  most present labels
popular_labels = counts.index[:10]
    

plt.figure(figsize=(15,8))
plt.barh(popular_labels, width=counts[:10],
        color=('brown','purple','red','green','orange','blue','black','crimson','yellow','cyan'))
plt.title('Top 10 Brand Distribution')
plt.xlabel('Percentage')
plt.ylabel('Brand')
plt.show()
```

Investigate Popularity

```python

# membuat array brand mobil
cars = np.array(df['Make'])
unique_car = np.unique(cars, return_index=True)[1]
cars = np.array([cars[index] for index in sorted(unique_car)])

# membuat array popularity
popularities = np.array(df['Popularity'])
unique_popularity = np.unique(popularities, return_index=True)[1]
popularities = np.array([popularities[index] for index in sorted(unique_popularity)])

# Get the indices sorted on popularities from highest to lowest
pop_index = np.argsort(popularities)[::-1]
pop_new_index = pop_index[0:20]
# Data range
total = pop_new_index.shape[0]


plt.figure(figsize=(20,8))
plt.title("Car Popularities")
plt.bar(range(total), popularities[pop_new_index], color="orange", align="center")
plt.xticks(rotation=90)
plt.xticks(range(total), cars[pop_new_index])
plt.xlim([-1, total])
plt.xlabel('Car Models')
plt.ylabel('Popularity')
plt.show()
```

# Investigate Mean MSRP by Engine Fuel Type and Vehicle Size

```python
fuel_Type=df.groupby(['Engine Fuel Type'])['MSRP'].mean().sort_values()
plt.figure(figsize=(15,4))
plt.bar(fuel_Type.index,fuel_Type.values, color=['red','green','blue','orange','purple'])
plt.title('Mean MSRP by Engine Fuel Type')
plt.show()
```

```python
VehicleSize=df.groupby(['Vehicle Size'])['MSRP'].mean()
```

```python
vehicle_size=df.groupby(['Vehicle Style'])['MSRP'].mean().sort_values()
plt.figure(figsize=(15,4))
plt.bar(vehicle_size.index,vehicle_size.values, color=['red','green','blue'])
plt.title('Mean MSRP by Vehicle Size')
plt.xticks(rotation=90)
plt.show()
```

# Most Expensive Ford and BMW Car

```python
model_bmw = df[df['Make'] == 'BMW']['Model']
```

```python
bmw_type = df.groupby(model_bmw)['MSRP'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(15,4))
plt.bar(bmw_type.index,bmw_type.values,color='orange')
plt.title('Most Expensive BMW Cars')
plt.ylabel('MSRP')
plt.xticks(rotation=90)
```

```python
model_ford = df[df['Make'] == 'Ford']['Model']
```

```python
ford_type = df.groupby(model_ford)['MSRP'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(15,4))
plt.bar(ford_type.index,ford_type.values, color='orange')
plt.title('Most Expensive Ford Cars')
plt.ylabel('MSRP')
plt.xticks(rotation=90)
```
