>  Handing Missing Data in 'Engine Fuel Type'

Investigation shows that all missing values have the same Car Company ('Make') and Car Model ('Model'). This case can be solved by using mode of Engine Fuel Type based on what model of this car has.

```python
car_df[(car_df['Make'] == 'Suzuki') & (car_df['Model'] == 'Verona')]
```

```python
# check the NaN value in dataset based 
car_df[car_df['Engine Fuel Type'].isna() == True]

#Create a function to fill tha NaN Values
def ApplyEngine(x):
    if(pd.isna(x['Engine Fuel Type'])):
        item =  x['Model']
        return car_df[car_df['Model'] == item]['Engine Fuel Type'].mode()[0]
    else:
        return x['Engine Fuel Type']
    
#fill NaN values with 
car_df['Engine Fuel Type'] = car_df[['Model','Engine Fuel Type']].apply(ApplyEngine,axis=1)
car_df.isna().sum()
```

> Handling Missing Values in Engine Cylinders

Some cars are electril fuel type so they dont have any cylinder in their engine. First, identify every Fuel type and the model to get some verification from some valid sources

```python
car_df[car_df['Engine Cylinders'].isna() == True][['Engine Fuel Type','Model']]
```

Those Models (Bolt EV, e-Golf, i-MiEV, RAV4 EV) are electric cars, so there is no cylinder configuration in their engine. Thus, this NaN Value can be added by 0. On The Other Hand some forums describe that RX-7 and RX-8 don't use Cylinders because their configuration system is using rotary engine, but some others say although they are using rotary engine but both Rx-7 and Rx-8 have 2 Cylinders. Hence, by putting 2 Cylinders sounds reasonable because of their fuel type.
<br>
source : https://www.proprofs.com/discuss/q/497458/a-mazda-rx8-has-how-many-cylinders

```python
#Creating Function to add number of Cylinders based on Engine Fuel Type
def ApplyCylinders(x):
    if(pd.isna(x['Engine Cylinders'])):
        if(x['Model'] == 'RX-8'):
            return 2
        elif(x['Model'] == 'RX-7'):
            return 2
        elif(x['Engine Fuel Type'] == 'electric'):
            return 0
    else:
        return x['Engine Cylinders']

#Using Apply Function to fill NaN Values
car_df['Engine Cylinders'] = car_df[['Model','Engine Cylinders','Engine Fuel Type']].apply(ApplyCylinders,axis=1)


car_df.isna().sum()
```

> Handling Missing Values in Engine Horse Power

```python
car_df[car_df['Engine HP'].isna() == True]
```

In Handling Horse Power, if the car is electrical then the NaN value in the Horse Power column is filled with the mean electrical car horsepower. However, if the car is non electrical, an average horsepower will be obtained from mean horsepower of number cylinders.

```python
#Creating Function to Add Engine HP

def ApplyHP(x):
    if(pd.isna(x['Engine HP'])):
        if(x['Engine Fuel Type'] == 'electric'):
            fuel_type = x['Engine Fuel Type']
            return round(car_df[car_df['Engine Fuel Type'] == fuel_type]['Engine HP'].mean(),1)
        else:
            cylinders_num = x['Engine Cylinders']
            return round(car_df[car_df['Engine Cylinders'] == cylinders_num]['Engine HP'].mean(),1)
    else:
        return x['Engine HP']

#Fill NaN Value of Engine HP 
car_df['Engine HP'] = car_df[['Engine HP','Engine Fuel Type','Engine Cylinders']].apply(ApplyHP,axis=1)

#Show data structure
car_df.isna().sum()
```

> Handling Missing Values in Missing Door

There are only 2 types of cars :
-Ferarri<br>
-Tesla<br>
https://www.cars.com/research/ferrari-ff-2013/ 2 Doors
<br>
https://www.tesla.com/models 4 Doors

```python
#Creating Function to Add Number of Doors
def ApplyDoor(x):
    if(pd.isna(x['Number of Doors'])):
        if(x['Model'] == 'FF'):
            return 2
        elif(x['Model'] == 'Model S'):
            return 4
    else:
        return x['Number of Doors']

#Fill NaN Values
car_df['Number of Doors'] = car_df[['Number of Doors','Model']].apply(ApplyDoor,axis=1)

car_df.isna().sum()
```
